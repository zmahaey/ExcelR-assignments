```python
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\zahee\\Desktop\\Data Science\\Project ExcelR\\telecommunications_churn.csv")
df
```

```python
df.drop(['voice_mail_plan','day_charge','evening_charge','night_charge', 'international_charge'],axis=1, inplace=True)
df
```

```python
# Splitting into X and Y
x= df.iloc[:,:-1]
Y= df.iloc[:,-1]
```

```python
# Standardizing the input variables
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
X=scale.fit_transform(x)
```

```python
# Splitting into test and train data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size=0.2, random_state=7)
```

# Grid Search CV

```python
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
clf=SVC()
param_grid=[{'kernel':['rbf'], 'gamma': [100,50,0.5,0.1,0.01,0.001,0.0001],
            'C':[15,14,13,12,11,10,0.01,0.001,0.0001]}]
gsv=GridSearchCV(clf,param_grid,cv=10)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_,gsv.best_score_
```

```python
clf=SVC(C=10, gamma=0.1)
clf.fit(x_train,y_train)
y_pred_train=clf.predict(x_train)
acc= accuracy_score(y_train,y_pred_train)*100
print("accuarcy=", acc)
print(confusion_matrix(y_train,y_pred_train))
```

```python
y_pred_test=clf.predict(x_test)
acc=accuracy_score(y_test,y_pred_test)*100
print("accuarcy=", acc)
print(confusion_matrix(y_test,y_pred_test))
```

### optimizing the model for imbalanced dataset

```python
weights={0:1,1:6}
clf=SVC(C=10, gamma=0.1, class_weight=weights)
clf.fit(x_train,y_train)
y_pred_train=clf.predict(x_train)
acc= accuracy_score(y_train,y_pred_train)*100
print("accuarcy=", acc)
print(confusion_matrix(y_train,y_pred_train))
```

```python
y_pred_test=clf.predict(x_test)
acc=accuracy_score(y_test,y_pred_test)*100
print("accuarcy=", acc)
print(confusion_matrix(y_test,y_pred_test))
```

### Thus, we get an accuracy of 91.15% after dealing with class imbalance
