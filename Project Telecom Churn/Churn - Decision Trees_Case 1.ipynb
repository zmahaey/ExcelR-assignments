```python
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\zahee\\Desktop\\Data Science\\Project ExcelR\\telecommunications_churn.csv")
df
```

```python
df.drop(['voice_mail_plan','day_charge','evening_charge','night_charge', 'international_charge'],axis=1, inplace=True)
df
```

```python
# Splitting into X and Y
X= df.iloc[:,:-1]
Y= df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
# Splitting into train and test data
x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size= 0.2, random_state=7)
```

```python
#Building model
model=DecisionTreeClassifier(criterion='entropy',max_depth=3)
model.fit(x_train,y_train)
```

```python
# Graphical representation
import matplotlib.pyplot as plt

fn= list(X.columns)
cn = ['churn','loyal']

fig,axes = plt.subplots(nrows=1, ncols=1, figsize=(4,4), dpi=300)

tree.plot_tree(model,feature_names=fn, class_names=cn, filled =True)
```

```python
preds=model.predict(x_test)
preds
```

```python
pd.crosstab(y_test,preds)
```

```python
import numpy as np
np.mean(y_test==preds)
```

# Applying Bagging ensemble technique to improve the accuracy

```python
from sklearn.ensemble import BaggingClassifier
cart=DecisionTreeClassifier()
model = BaggingClassifier(base_estimator=cart,n_estimators=1000,random_state=7)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

# Applying Random Forest ensemble technique

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold

kfold=StratifiedKFold(n_splits=10, random_state=7, shuffle=True)
model=RandomForestClassifier(n_estimators=100, max_features=3)
score = cross_val_score(model,X,Y,cv=kfold)
print(score.mean())
```

# Applying Ada Boosting ensemble technique

```python
from sklearn.ensemble import AdaBoostClassifier
kfold=StratifiedKFold(n_splits=10,random_state=7, shuffle=True)
model= AdaBoostClassifier(n_estimators=10,random_state=7)
score=cross_val_score(model,X,Y,cv=kfold)
print(score.mean())
```

# Applying Stacking ensemble technique

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier

kfold=StratifiedKFold(n_splits=10,random_state=7, shuffle = True)
```

```python
# Create the sub models
estimators=[]
model1=LogisticRegression(max_iter=500)
estimators.append(('logistic',model1))
model2= DecisionTreeClassifier()
estimators.append(('cart',model2))
model3= SVC()
estimators.append(('svm',model3))
```

```python
#Create the ensemble model
import warnings
warnings.filterwarnings("ignore")
ensemble =VotingClassifier(estimators)
results=cross_val_score(ensemble,X,Y,cv=kfold)
print(results.mean())
```

# Thus, we see that Random Forest classifier gives the highest accuracy of 97.87%
